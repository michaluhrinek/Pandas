```python
# Convert to datetime and extract date part
df['LAST UPDATE'] = pd.to_datetime(df['LAST UPDATE'], errors='coerce').dt.date
```

```python
df['geo'] = df['geo'].str.strip().str.upper()
```

```python
#Get only columns of a choise plus calculation of the average of gas price per year.
```

```python
# Ensure OBS_VALUE is numeric
df['OBS_VALUE'] = pd.to_numeric(df['OBS_VALUE'], errors='coerce')

# Remove rows with missing or zero OBS_VALUE
df = df[df['OBS_VALUE'].notnull() & (df['OBS_VALUE'] > 0)]

# Extract year from TIME_PERIOD (assumes format like '2021-S1')
df['YEAR'] = df['TIME_PERIOD'].str[:4]

# Group by country and year, calculate mean
df_new = df.groupby(['geo', 'YEAR'])['OBS_VALUE'].mean().reset_index()

# Rename the column for clarity
df_new.rename(columns={'OBS_VALUE': 'AVG_OBS_VALUE'}, inplace=True)

# Show result
print(df_new.head())
```

```python
import matplotlib.pyplot as plt

# Find the latest year in the df_new DataFrame
latest_year = df_new['YEAR'].max()

print(f"Plotting average gas price by country for the latest available year: {latest_year}")

# Filter for the latest year
df_latest_year = df_new[df_new['YEAR'] == latest_year]

# Sort by price (optional, for better visualization)
df_latest_year = df_latest_year.sort_values('AVG_OBS_VALUE', ascending=False)

# Plot the bar chart
plt.figure(figsize=(15,6))
plt.bar(df_latest_year['geo'], df_latest_year['AVG_OBS_VALUE'], color='skyblue')
plt.title(f'Average Gas Price by Country in {latest_year}')
plt.ylabel('Average Gas Price (Euro per GJ)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Calculate average price per country over the entire period
country_avg = df.groupby('geo')['OBS_VALUE'].mean().sort_values(ascending=False)

# Get top 5 countries
top_5_countries = country_avg.head(5).index.tolist()

# Filter data for top 5 countries
df_top5 = df[df['geo'].isin(top_5_countries)].copy()

# Group by country and year, calculate mean price
df_chart = df_top5.groupby(['geo', 'TIME_PERIOD'])['OBS_VALUE'].mean().reset_index()

# Create the line chart
plt.figure(figsize=(12, 8))

# Plot line for each country
for country in top_5_countries:
    country_data = df_chart[df_chart['geo'] == country]
    plt.plot(country_data['TIME_PERIOD'], country_data['OBS_VALUE'],
             marker='o', linewidth=2, label=country)

# Customize the chart
plt.title('Top 5 European Countries - Gas Prices Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Gas Price', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the chart
plt.show()

# Optional: Print the top 5 countries and their average prices
print("Top 5 Countries by Average Gas Price:")
for i, (country, avg_price) in enumerate(country_avg.head(5).items(), 1):
    print(f"{i}. {country}: {avg_price:.2f}")
```

```python
# 6. GEOGRAPHICAL ANALYSIS
print("\n\n6. GEOGRAPHICAL ANALYSIS")
print("-" * 40)

# Top countries by average price
country_stats = df.groupby('geo').agg({
    'OBS_VALUE': ['mean', 'median', 'std', 'count']
}).round(2)
country_stats.columns = ['Mean_Price', 'Median_Price', 'Price_Std', 'Record_Count']
country_stats = country_stats.sort_values('Mean_Price', ascending=False)

print("Top 10 Countries by Average Price:")
print(country_stats.head(10))

print("\nBottom 10 Countries by Average Price:")
print(country_stats.tail(10))

# Visualize geographical patterns
plt.figure(figsize=(16, 12))
```

```python
# Top 15 countries by average price
plt.subplot(2, 2, 1)
top_countries = country_stats.head(15)
top_countries['Mean_Price'].plot(kind='bar', color='coral')
plt.title('Top 15 Countries by Average Gas Price')
plt.ylabel('Average Price')
plt.xticks(rotation=45)

# Price volatility by country (top 15)
plt.subplot(2, 2, 2)
top_volatile = country_stats.nlargest(15, 'Price_Std')
top_volatile['Price_Std'].plot(kind='bar', color='orange')
plt.title('Top 15 Countries by Price Volatility')
plt.ylabel('Price Standard Deviation')
plt.xticks(rotation=45)

# Records per country (top 15)
plt.subplot(2, 2, 3)
top_records = country_stats.nlargest(15, 'Record_Count')
top_records['Record_Count'].plot(kind='bar', color='lightblue')
plt.title('Top 15 Countries by Number of Records')
plt.ylabel('Record Count')
plt.xticks(rotation=45)

# Scatter plot: Mean vs Volatility
plt.subplot(2, 2, 4)
plt.scatter(country_stats['Mean_Price'], country_stats['Price_Std'], alpha=0.6, color='purple')
plt.xlabel('Mean Price')
plt.ylabel('Price Volatility (Std Dev)')
plt.title('Price Level vs Volatility by Country')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
